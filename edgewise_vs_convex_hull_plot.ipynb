```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scales = np.load('data_comparison/scales.npy')
micp = np.load('data_comparison/cost_2norm_micp.npy')
micp_squared = np.load('data_comparison/cost_2norm_squared_micp.npy')
edgewise = np.load('data_comparison/cost_2norm_edgewise.npy')
edgewise_squared = np.load('data_comparison/cost_2norm_squared_edgewise.npy')
ch = np.load('data_comparison/cost_2norm_ch.npy')
ch_squared = np.load('data_comparison/cost_2norm_squared_ch.npy')
proposed = np.load('data_comparison/cost_2norm_proposed.npy')
proposed_squared = np.load('data_comparison/cost_2norm_squared_proposed.npy')
```

```python
def micp_vs_relaxation(micp, proposed, edgewise, ch):
    plt.plot(scales, micp, label='MICP', linestyle='-', linewidth=2)
    plt.plot(scales, proposed, label='Proposed formulation', linestyle='--', linewidth=2)
    plt.plot(scales, edgewise, label='Edge-by-edge formulation', linestyle='-.', linewidth=2)
    plt.plot(scales, ch, label='Convex-hull formulation', linestyle=':', linewidth=2)
    plt.xlabel(r'Scale factor $r$')
    plt.ylabel('Cost')
    plt.xlim([scales[0], scales[-1]])
    plt.xscale('log')
    plt.grid(1)
    plt.legend(loc=3)
```

```python
plt.figure(figsize=(5, 3))
micp_vs_relaxation(micp, proposed, edgewise, ch)
```

```python
plt.figure(figsize=(5, 3))
micp_vs_relaxation(micp_squared, proposed_squared, edgewise_squared, ch_squared)
```
